## Linear Regression

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
import tensorflow
```

```python
# X_train, X_test, y_train, y_test = train_test_split(
# X_train = data_2023['open'][:187]
## Trying to predict the open price of the stock using linear regression
X = data_2023['open'].values.reshape(-1, 1)
y = data_2023['open'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)

model = LinearRegression()
model.fit(X_train, y_train)
```

```python
# Predicting the test set results
y_pred = model.predict(X_test)

# Visualising the Training set results
plt.figure(figsize=(12, 6))
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.title('Open vs Close (Training set)')
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()

# Visualising the Test set results
plt.figure(figsize=(12, 6))
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.title('Open vs Close (Test set)')
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()
```

```python
# Calculating the error
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```

```python
plotting_data = [*X_train[:200].flatten(), *y_pred.flatten()]
data_2023['predicted_linear'] = plotting_data
plot_trends('predicted_linear', 'Predicted Open Prices using Linear Regression')
plot_trends('open', 'Actual Open Prices')
```

## ANN

```python
import tensorflow as tf
import tensorboard as tb
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Dense(1, input_shape=(1,), activation='relu'),
    tf.keras.layers.Dense(1, activation='relu'),
    tf.keras.layers.Dense(1, activation='relu'),
])

model.compile(optimizer='adam', loss='mean_squared_error', metrics=[tf.keras.metrics.RootMeanSquaredError()])

model.fit(X_train, y_train, epochs=100)
```

```python
# Predicting the test set results
y_pred = model.predict(X_test)

# Visualising the Training set results
plt.figure(figsize=(12, 6))
plt.scatter(X_train, y_train, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.title('Open vs Close (Training set)')
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()

# Visualising the Test set results
plt.figure(figsize=(12, 6))
plt.scatter(X_test, y_test, color='red')
plt.plot(X_train, model.predict(X_train), color='blue')
plt.title('Open vs Close (Test set)')
plt.xlabel('Open')
plt.ylabel('Close')
plt.show()
```
